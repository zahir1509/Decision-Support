```python
import string
import nltk
import csv
nltk.download('punkt')
nltk.download('vader_lexicon')
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
```

```python
#This array contains the polarity score for each hotel
scoreForEachHotel = [['Negative', 'Neutral', 'Positive','Compound']]
```

```python

def sentiment_analyse(sentiment_text):
    score = SentimentIntensityAnalyzer().polarity_scores(sentiment_text)
    print("-------------score-------------")
    print(score)

    # Appending the score in the array
    
    scoreForEachHotel.append([score['neg'],score['neu'],score['pos'],score['compound']])

```

```python

df = pd.read_csv('reviews_test.csv')
reviewcol = df.Review

for text in reviewcol:
  text = str(text)
  lower_case = text.lower()
  cleaned_text = lower_case.translate(str.maketrans('', '', string.punctuation))
  print(cleaned_text)
  sentiment_analyse(cleaned_text)


```

```python
sfeh = pd.DataFrame(scoreForEachHotel)
sfeh
```

```python
sfeh.columns = sfeh.iloc[0]
sfeh
```

```python
sfeh=sfeh.iloc[1:]
sfeh= sfeh.reset_index()
sfeh
```

```python
sfeh = sfeh.drop(['index'], axis=1)
sfeh
```
