```python
sg_loc = locator.geocode("Singapore")
bk_loc = locator.geocode("Bangkok")
kl_loc = locator.geocode("Kuala Lumpur")
```

```python
map_sg = folium.Map(
    location=[sg_loc.latitude, sg_loc.longitude],
    tiles='cartodbpositron',
    zoom_start=12,
)

map_bk = folium.Map(
    location=[bk_loc.latitude, bk_loc.longitude],
    tiles='cartodbpositron',
    zoom_start=12,
)

map_kl = folium.Map(
    location=[kl_loc.latitude, kl_loc.longitude],
    tiles='cartodbpositron',
    zoom_start=12,
)

```

```python
sortedbyrev2 = sortedbyrev.dropna()
sortedbyrev2.apply(lambda row:folium.Marker(location=[row["latitude"], row["longitude"]], popup=(row['NandC'] + ", Rating: " + str(row['Overall Rating']) + ", Polarity: " + str(row['Compound'])[:6])).add_to(map_sg), axis=1)
sortedbyrev2.apply(lambda row:folium.Marker(location=[row["latitude"], row["longitude"]], popup=(row['NandC'] + ", Rating: " + str(row['Overall Rating']) + ", Polarity: " + str(row['Compound'])[:6])).add_to(map_bk), axis=1)
sortedbyrev2.apply(lambda row:folium.Marker(location=[row["latitude"], row["longitude"]], popup=(row['NandC'] + ", Rating: " + str(row['Overall Rating']) + ", Polarity: " + str(row['Compound'])[:6])).add_to(map_kl), axis=1)

sortedbypol2 = sortedbypol.dropna()
sortedbypol2.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]], popup=(row['NandC'] + ", Rating: " + str(row['Overall Rating']) + ", Polarity: " + str(row['Compound'])[:6])).add_to(map_sg), axis=1)
sortedbypol2.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]], popup=(row['NandC'] + ", Rating: " + str(row['Overall Rating']) + ", Polarity: " + str(row['Compound'])[:6])).add_to(map_bk), axis=1)
sortedbypol2.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]], popup=(row['NandC'] + ", Rating: " + str(row['Overall Rating']) + ", Polarity: " + str(row['Compound'])[:6])).add_to(map_kl), axis=1)
```
